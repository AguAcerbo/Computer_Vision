```python
#Si queremos que las imágenes sean mostradas en una ventana emergente quitar el inline
#%matplotlib inline

# OpenCV-Python utiliza NumPy para el manejo de imágenes
import numpy as np
# cv2 es el módulo python para acceder a OpenCV 
import cv2 as cv
# Usamos las poderosas herramientas de graficación de matplotlib para mostrar imágenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt
```

```python
def LBP_basico(direccion):
    '''
    Esta funcion recibe la direccion de una imagen y devuelve la imagen en colo, la imagen en escalas de grises y
    la caracteristica LBP basica, considerando los 8 vecinos inmediatos de cada pixel.
    '''
    
    imagen_color = cv.imread(direccion)

    imagen_gray = cv.cvtColor(imagen_color,cv.COLOR_BGR2GRAY)
    img_lbp = np.zeros((imagen_gray.shape[0]-2,imagen_gray.shape[1]-2))

    for p_row in range(1, imagen_gray.shape[0]-1):
        for p_column in range(1, imagen_gray.shape[1]-1):
            coordenadas = np.array([[p_row, p_column - 1],
                                    [p_row + 1, p_column - 1],
                                    [p_row + 1, p_column],
                                    [p_row + 1, p_column + 1],
                                    [p_row, p_column + 1],
                                    [p_row - 1, p_column + 1],
                                    [p_row - 1, p_column],
                                    [p_row - 1, p_column - 1]])

            validos = (imagen_gray[coordenadas[:,0],coordenadas[:,1]] > imagen_gray[p_row,p_column])
            img_lbp[p_row-1,p_column-1] = sum(validos * np.power(2,range(8)))
    
    return imagen_color, imagen_gray, img_lbp
```

```python
# No es con barras
def histogram_lbp(imagen1, imagen2):
    hist1, bins1 = np.histogram(imagen1.ravel(),256,[0,256])
    hist2, bins2 = np.histogram(imagen2.ravel(),256,[0,256])

    fig, ax = plt.subplots(1,2,figsize =(20, 5))
    ax[0].plot(hist1)
    ax[1].plot(hist2)
    fig2, ax2 = plt.subplots(figsize =(20, 5))
    ax2.plot(hist1)
    ax2.plot(hist2)
    return None
```

```python
def histogram_lbp2(imagen1, imagen2):
    hist1, bins1 = np.histogram(imagen1.ravel(),256,[0,256])
    hist2, bins2 = np.histogram(imagen2.ravel(),256,[0,256])
    
    x1=range(len(hist1))
    x2=range(len(hist2))
    fig, ax = plt.subplots(1,2,figsize =(20, 5))
    plt.xticks([0, 50, 100, 150, 200, 255],[0, 50, 100, 150, 200, 255])
    ax[0].bar(x1, hist1, width = 1, align='edge')
    ax[1].bar(x1, hist2, width = 1, align='edge')
    fig2, ax2 = plt.subplots(figsize =(20, 10))
    ax2.bar(x1, hist1, width = 1, align='edge')
    ax2.bar(x1, hist2, width = 1, align='edge')
    
    hist_comparation(imagen1, imagen2)
    
    return None
```

```python
def graficar_histograma(img):
    hist, bins = np.histogram(img.ravel(),200,[0,256])
    fig, ax = plt.subplots(figsize =(20, 10))
    
    x=range(len(hist))
    plt.xticks([0, 50, 100, 150, 200, 255],[0, 50, 100, 150, 200, 255])
    ax.bar(x, hist, align='edge')

    return None
```

```python
def hist_comparation(img1, img2):
    img11 = img1.copy()
    img22 = img2.copy()
    
    img1_f32 =  np.float32(img11)
    hist_1 = cv.calcHist(img1_f32, [0], None, [256], [0,256], accumulate=False)
    img2_f32 =  np.float32(img22)
    hist_2 = cv.calcHist(img2_f32, [0], None, [256], [0,256], accumulate=False)
    
    img1_img2_corr = cv.compareHist(hist_1, hist_2, cv.HISTCMP_CORREL)
    img1_img2_chi2 = cv.compareHist(hist_1, hist_2, cv.HISTCMP_CHISQR)
    img1_img2_intersect = cv.compareHist(hist_1, hist_2, cv.HISTCMP_INTERSECT)
    img1_img2_bhattacharyya = cv.compareHist(hist_1, hist_2, cv.HISTCMP_BHATTACHARYYA)
    
    print(f'Correlation: {img1_img2_corr}')
    print(f'Chi-Square: {img1_img2_chi2}')
    print(f'Intersection: {img1_img2_intersect}')
    print(f'Bhattacharyya distance: {img1_img2_bhattacharyya}')
    
    return None
```
